По заданию дана функция f(x) = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30

Преподаватель на семинаре разрешил использовать функцию f(x) = -5*x**2 + 10*x - 150

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot as plt
init_printing()

x = Symbol('x')
fx = -5*x**2 + 10*x - 150
```

**1. Определить корни** 

```python
x1, x2 = solve(fx)
print(f'Первый корень = {round(x1.evalf(), 2)} \nВторой корень = {round(x2.evalf(), 2)}')
```

**2. Найти интервалы, на которых функция возрастает**

```python
solve(diff(fx)>0)
```

**3. Найти интервалы, на которых функция убывает** 

```python
solve(diff(fx)<0)
```

**4. Построить график**

```python
p = plt(fx, (x,-8,10))
```

**5. Вычислить вершину**

```python
solve(diff(fx))
```

**6. Определить промежутки, на котором f > 0**

```python
solve(fx>0)
```

**7. Определить промежутки, на котором f < 0**

```python
solve(fx<0)
```
